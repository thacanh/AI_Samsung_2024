```python
import pandas as pd
data=pd.read_csv('/content/drive/MyDrive/DataAISamsung2024/AB_NYC_2019 (1).csv')
data
```

```python
# 1. váº½ biá»ƒu Ä‘á»“ cá»™t cho neighborhood group
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='neighbourhood_group')
plt.title('Sá»‘ lÆ°á»£ng cÃ¡c phÃ²ng theo tá»«ng khu vá»±c lá»›n')
plt.xlabel('Khu vá»±c')
plt.ylabel('Sá»‘ lÆ°á»£ng phÃ²ng')
plt.show()
```

```python
#2
plt.figure(figsize=(12, 8))
sns.histplot(data=data, x='neighbourhood', kde=False, bins=50)
plt.title('Sá»‘ lÆ°á»£ng cÃ¡c phÃ²ng theo tá»«ng khu vá»±c nhá»')
plt.xlabel('Khu vá»±c')
plt.ylabel('Sá»‘ lÆ°á»£ng phÃ²ng')
plt.xticks(rotation=45)
plt.show()
```

```python
#3
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='neighbourhood_group', y='availability_365')
plt.title('Má»‘i liÃªn há»‡ giá»¯a khu vá»±c vÃ  sá»‘ ngÃ y cÃ³ phÃ²ng trá»‘ng')
plt.xlabel('Khu vá»±c')
plt.ylabel('Sá»‘ ngÃ y')
plt.xticks(rotation=45)
plt.show()
```

```python
#4
plt.figure(figsize=(10, 8))
sns.scatterplot(data=data, x='longitude', y='latitude', hue='neighbourhood_group', alpha=0.5, palette='viridis')
plt.title('Map')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Neighbourhood Group')
plt.show()
```

```python
#5
numeric_data = data.select_dtypes(include='number')
plt.figure(figsize=(14, 12))
corr_matrix = numeric_data.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Heatmap')
plt.show()
```
